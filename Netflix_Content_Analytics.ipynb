```python
import pandas as pd

df = pd.read_csv('netflix_titles.csv.zip')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
print("Rows and Columns:", df.shape)
```

```python
content_type = df['type'].value_counts()

print(content_type)
```

```python
import matplotlib.pyplot as plt

content_type.plot(kind='bar')

plt.title('Movies vs TV Shows on Netflix')
plt.xlabel('Content Type')
plt.ylabel('Number of Titles')

plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)

print(top_countries)
```

```python
top_countries.plot(kind='bar', figsize=(10,5))

plt.title('Top 10 Content Producing Countries')
plt.xlabel('Country')
plt.ylabel('Number of Titles')

plt.show()
```

```python
genres = df['listed_in'].str.split(', ', expand=True).stack()

top_genres = genres.value_counts().head(10)

print(top_genres)
```

```python
top_genres.plot(kind='bar', figsize=(10,5))

plt.title('Top 10 Netflix Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')

plt.xticks(rotation=45)
plt.show()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed')
```

```python
df['year_added'] = df['date_added'].dt.year
```

```python
yearly_content = df['year_added'].value_counts().sort_index()

print(yearly_content)
```

```python
yearly_content.plot(kind='line', marker='o', figsize=(10,5))

plt.title('Netflix Content Added Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Titles Added')

plt.grid(True)

plt.show()
```
